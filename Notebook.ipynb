```python
import pandas as pd
```

```python
df=pd.read_csv('IPL.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

Option A: Drop rows where critical identifier columns are missing
df = df.dropna(subset=['critical_column'])

 Option B: Fill missing numerical data with the median value
 We use median instead of mean because it is less sensitive to extreme outliers.
 numerical_cols = df.select_dtypes(include=['number']).columns
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

 Option C: Fill missing categorical data with a placeholder string
df['categorical_column'] = df['categorical_column'].fillna('Unknown')

```python
df.drop_duplicates
```

```python
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df['runs'] = pd.to_numeric(df['runs_total'], errors='coerce')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

Example A: Converting a column to integer or float if read as string
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

Example B: Converting dates stored as strings into proper datetime format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['strike_rate'] = (df['runs_total'] / df['balls_faced']) * 100
print(df['strike_rate'])
```

```python
df.to_csv('cleaned_data.csv')
```
